# Basic OpenCV Functions

**Importing required Librariess**

```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Uploading an image to work on

```
from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```
image_name="test2.jpg"
```

**Opening Images using Opencv**

```
image=cv2.imread(str(image_name))
```

Display captured image

```
f = plt.figure(figsize=(6,6))
print("The captured image is shown below")
plt.imshow(image)
```

**Doesn't look like the image you uploaded??**

By default openCV accepts images in BGR format

![image channels](https://www.kdnuggets.com/wp-content/uploads/image-data-analysis.jpg)

**Converting the image back to RGB format**

```
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

f = plt.figure(figsize=(6,6))
print("The converted RGB image")
plt.imshow(image)
```

Better?


**How to resize this image to some required shape?**

```
image_resized=cv2.resize(image,(224,224))

f = plt.figure(figsize=(6,6))
print("Resized image")
plt.imshow(image_resized)
```

**Extracting grayscale images**

```
gray_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

f = plt.figure(figsize=(6,6))
print("grayscale image")
plt.imshow(gray_scale_image)
```

Not Grayscale?

This is because OpenCV is trying to draw 3 channels whereas only one channel is present

![grayscale](https://upload.wikimedia.org/wikipedia/commons/thumb/3/33/Beyoglu_4671_tricolor.png/400px-Beyoglu_4671_tricolor.png)

```
f = plt.figure(figsize=(6,6))

print("grayscale image")
plt.imshow(gray_scale_image, cmap='gray')
```

**Displaying results on images**

```
draw_image=image.copy()

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(draw_image,'Result: apple',(200,60), font, 1.5,(255,0,0),2,cv2.LINE_AA)

f = plt.figure(figsize=(6,6))

print("Text on image")
plt.imshow(draw_image)
```

**Draw Polygons**

```
cv2.rectangle(draw_image,(175,135),(490,415),(0,255,0),5)
f = plt.figure(figsize=(6,6))

print("Rectangle on image")
plt.imshow(draw_image)
```

```
image_draw2=image.copy()

#drawing circle
cv2.circle(image_draw2,(200,150), 100, (255,0,255), 4)

#drawing polygon
pts = np.array([[500,250],[550,300],[550,350],[500,350],[520,300]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image_draw2,[pts],True,(0,255,255),4)

f = plt.figure(figsize=(6,6))

print("Other shapes")
plt.imshow(image_draw2)
```

# Other OPenCV functions

**edge Detection**

```
frame = image.copy()

frame=cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)

hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) 
      
# define range of red color in HSV 
lower_red = np.array([30,150,50]) 
upper_red = np.array([255,255,180]) 
      
# create a red HSV colour boundary and  
# threshold HSV image 
mask = cv2.inRange(hsv, lower_red, upper_red) 
  
# Bitwise-AND mask and original image 
res = cv2.bitwise_and(frame,frame, mask= mask) 

edges = cv2.Canny(frame,100,200)

f = plt.figure(figsize=(6,6))

print("Other shapes")
plt.imshow(edges, cmap="gray")
```
